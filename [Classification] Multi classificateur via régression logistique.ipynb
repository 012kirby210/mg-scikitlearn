```python
%matplotlib inline

# On utilise la classification binaire offerte par la régression logistique mais on change l'interprétation 
# d'une des deux classes en "tout sauf l'autre classe"

import matplotlib.pyplot as plt
import mglearn

from sklearn.datasets import make_blobs

X,y = make_blobs()
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Paramètre 1")
plt.ylabel("Paramètre 2")
plt.legend(["Classe 1","Classe 2", "Classe 3"])
```

```python
# En utilisant un classificateur SVM, on constate la présence de 3 vecteurs pour les 3 classes,
# chaque vecteur détermine une classe et l'absence de cette dernière

from sklearn.svm import LinearSVC
linear_svm = LinearSVC(max_iter=10000).fit(X,y)
print("Forme des coeffs : ", linear_svm.coef_.shape)
print("Forme de l'ordonnée à l'origine : ", linear_svm.intercept_.shape)
```

```python
# Le modèle linéraire pour chaque cible est 
# y = ax1 + bx2 + c
# La frontière est obtenue pour l'équation caractéristique
# 0 = ax1 + bx2 + c ( <=> ) x2 = - (ax1 + c) / b

import numpy as np
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15.0, 15.0)
for coef, intercept, color in zip( linear_svm.coef_, linear_svm.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line*coef[0] + intercept)/coef[1], c=color)
plt.ylim(-14,10)
plt.xlim(-10,8)
plt.xlabel("Paramètre 1")
plt.ylabel("Paramètre 2")
plt.legend(["Classe 1", "Classe 2", "Classe 3", "Frontière C1", "Frontière C2", "Frontière C3"], 
          loc=(1.01, 0.3))
```

```python
# L'attribution d'une classe à un point qui est dans la surface de l'intersection est effectuée 
# de manière steochiométrique
mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15,15)
for coef, intercept, color in zip(linear_svm.coef_,linear_svm.intercept_,mglearn.cm3.colors):
    plt.plot(line, -(line*coef[0]+intercept)/coef[1], c=color)
plt.legend(["Classe 1", "Class 2", "Classe 3","Frontière C1", "Frontière C2", "Frontière C3"], 
          loc=(1.01, 0.3))
plt.xlabel("Paramètre 1")
plt.ylabel("Paramètre 2")
```
