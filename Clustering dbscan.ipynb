```python
%matplotlib inline
# Le clustering Density Based Spatial Clustering of Application with Noise est une méthode qui détecte le nombre
# de cluster par lui_même en fonctiond de ses paramètres d'apprentissage : 
# 1 - le nombre minimal de points par cluster
# 2 - une distance de seuil qui détermine les frontières des clusters

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

X,y = make_blobs(random_state=0,n_samples=12)
dbscan = DBSCAN()
clusters = dbscan.fit_predict(X)
print("Affiliation des points par cluster {}".format(clusters))
```

```python
# Tous les points précédens sont considérés comme du bruit
# Les paramètres par défaut de l'algo de clustering sont inadaptés au jeu d'échantillon
import mglearn
mglearn.plots.plot_dbscan()
```

```python
# Par défaut, l'aglo trouve les deux clusters de point pour le jeu d'échatillon two moons
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=200,noise=0.05, random_state=0)

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

dbscan = DBSCAN()
clusters = dbscan.fit_predict(X_scaled)
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=clusters, cmap=mglearn.cm2, s=60)
plt.xlabel("Paramètre 0")
plt.ylabel("Paramètre 1")
```

```python
# Par contre si on modifie un peu la valeur de eps
from matplotlib.colors import ListedColormap

dbscan = DBSCAN(eps=0.2)
clusters = dbscan.fit_predict(X_scaled)
color_map=ListedColormap(["#8ecae6","#219ebc","#023047","#ffb703","#fb8500","#8338ec"])
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=clusters, cmap=color_map, s=60)
plt.xlabel("Paramètre 0")
plt.ylabel("Paramètre 1")
# On trouve un nombre incongru de clusters
```
