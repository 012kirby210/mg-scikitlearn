```python
%matplotlib inline
# Application d'algos de regression linéaire à la classification

import matplotlib.pyplot as plt
import mglearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X,y = mglearn.datasets.make_forge()
fig, axes = plt.subplots(1,2, figsize=(10,3))
for model,ax, in zip([LinearSVC(), LogisticRegression()],axes) :
    classifier = model.fit(X,y)
    mglearn.plots.plot_2d_separator(classifier, X, fill=False,eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{}".format(classifier.__class__ .__name__))
    ax.set_xlabel("Paramètre 0")
    ax.set_ylabel("Paramètre 1")
axes[0].legend()
```
